```python
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats
```

```python
x = [2,4,6,8,12,13,15,18,20,25,27,28,25,26,32,34,36,38]
y = [1, 20, 5, 21, 10, 27, 13, 49, 18, 60, 18, 62, 70, 87, 55, 77, 73, 80]
```

```python
# x_train = x[:-5]
# y_train = y[:-5]
# x_test = x[-5:]
# y_test = y[-5:]

x_train = [2,4,8,12,15,18,20,27,28,25,32,34,36]
y_train = [1, 20, 21, 10, 13, 49, 18, 18, 62, 70, 55, 77, 73]

x_test = [6,13,25,26,38]
y_test = [5,27,60,87,80]
```

```python
slope, intercept, r, p, std_err = stats.linregress(x_train, y_train)
mean = np.mean(y_train)
```

```python
def get_predicted_y(x_value):
    return intercept + slope * x_value

def get_mean_y(x_value):
    return mean

y_predicted = list(map(get_predicted_y, x_train))
y_mean = list(map(get_mean_y, x_train))

```

```python
plt.scatter(x_train, y_train)
plt.plot(x_train, y_predicted)
plt.plot(x_train, y_mean)
plt.scatter(x_test, y_test, color='red')
plt.show()
```
